Функции, например, можно задавать так. Зададим функцию $U(a, b)$

```python
def U(a, b):
    r = a + rd.random() * (b - a)
    return r
```

Вызов функции происходит как обычно. Возвращается случайное число из интервала от 2 до 5, предполагая, что случайная величина распределена равномерно на этом отрезке.

```python
U(2, 5)
```

Генерируем 100 случайных точек распределения $U(2, 5)$ и отобразим на гистограмме

```python
arr = np.zeros(100) 
for i in range(100):
    arr[i] = U(2, 5)
n,bins,patches = plt.hist(arr, range = (0.5, 6.5),bins = 20)
```
